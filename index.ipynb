## Step 4: Run a simple regression in Statsmodels with TV as a predictor

## Step 5: Get Regression Diagnostics Summary

```python
f = 'sales~TV'
model = ols(formula = f, data = df).fit()
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

The data is heteroscedastic.

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

The data is clearly normal.

## Step 9: Repeat the above for radio and record your observations

```python
f_radio = 'sales~radio'
model_radio = ols(formula = f_radio, data = df).fit()
model_radio.summary()
```

```python
X_new_radio = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
Y_vector_radio = model_radio.predict(X_new_radio)


df.plot(kind='scatter', x='radio', y='sales')
plt.plot(X_new_radio, Y_vector_radio)
plt.show()
```

```python
residuals = model_radio.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

Neither TV nor radio is a strong predictor for sales. Newspaper is has no chance of being a predictor for sales since it doesn't meet a test for linear data. At the same time, neither has high consistency of variability across values of either independent variable. The r-squared for TV is 0.612 and the r-squared for radio is 0.332. The QQ plot for TV is strong, much less so for radio.
